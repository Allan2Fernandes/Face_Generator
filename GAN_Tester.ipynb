```python
import matplotlib
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = []
for epoch in range(1, 21):
    model = tf.keras.models.load_model("Saved_models/ModelEpoch{}".format(epoch*5))
    models.append(model)
```

```python
def display_image(image_array):
    image_array = image_array*0.5 + 0.5
    plt.imshow(image_array)
    plt.show()
```

```python
for index, model in enumerate(models):
    generator, discriminator = model.layers
    prediction = 0
    generated_image = None
    while prediction < 0.7:
        noise = tf.random.normal(shape=(1, 128))
        generated_image = generator(noise)
        #display_image(generated_image[0])
        prediction = discriminator(generated_image)
        #print(prediction)
        pass
    print("Epochs train = {}".format((index+1)*5))
    print("Prediction certainty = {0:f}%".format(tf.get_static_value(prediction[0][0]*100)))
    display_image(generated_image[0])
    
```
