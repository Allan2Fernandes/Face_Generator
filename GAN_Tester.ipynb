```python
import matplotlib
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
base_path = "Saved_models"
model_names = os.listdir(base_path)
```

```python
model_paths = []
for model_name in model_names:
    path = os.path.join(base_path, model_name)
    model_paths.append(path)
    pass
```

```python
models = []
for path in model_paths:
    model = tf.keras.models.load_model(path)
    models.append(model)
```

```python
def display_image(image_array):
    image_array = image_array*0.5 + 0.5
    plt.imshow(image_array)
    plt.show()
```

```python
for model, model_name in zip(models, model_names):
    generator, discriminator = model.layers
    prediction = 0
    generated_image = None
    while prediction < 0.5:
        noise = tf.random.normal(shape=(1, 128))
        generated_image = generator(noise)
        #display_image(generated_image[0])
        prediction = discriminator(generated_image)
        #print(prediction)
        pass
    print("Model: {}".format(model_name))
    print("Prediction certainty = {0:f}%".format(tf.get_static_value(prediction[0][0]*100)))
    display_image(generated_image[0])
    
```
