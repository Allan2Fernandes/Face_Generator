```python
import matplotlib
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = []
for epoch in range(1, 5):
    model = tf.keras.models.load_model("Saved_models/ModelEpoch {}".format(epoch*5))
    models.append(model)
```

```python
def display_image(image_array):
    image_array = image_array*0.5 + 0.5
    plt.imshow(image_array)
    plt.show()
```

```python
def generated_images():
    images = []
    for model in models:
        noise = tf.random.normal(shape=(1, 128))
        generated_tensor = model(noise)
        image = generated_tensor[0]
        images.append(image)
        pass

    for step, image in enumerate(images):
        print("model epochs: {}".format((step+1)*5))
        display_image(image)
        pass
    
```

```python
generated_images()
```
